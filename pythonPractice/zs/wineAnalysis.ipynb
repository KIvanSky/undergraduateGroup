## 葡萄酒质量分析

### 数据预处理

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
FILE_PATH = 'winequality-red.csv'
```

```python
file = pd.read_csv(FILE_PATH) #读取红酒质量数据文件
```

```python
wine_index = {'fixed acidity': '固定酸',
           'volatile acidity': '挥发性酸',
           'citric acid': '柠檬酸',
           'residual sugar': '残留糖分',
           'chlorides': '氯化物',
           'free sulfur dioxide': '游离二氧化硫',
           'total sulfur dioxide': '总二氧化硫',
           'density': '浓厚度',
           'pH': '氢离子浓度',
           'sulphates': '硫酸盐',
           'alcohol': '酒精度',
           'quality': '质量'} 
file.rename(columns = wine_index, inplace = True)
file.head(10) #共1599条数据，这里显示10条
```

```python
file.info() # 查看数据缺损情况
```

```python
file.describe()
```
